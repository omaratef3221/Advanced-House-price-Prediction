```python
def remove_outliers(column):
    Q1 = df[column].quantile(0.4)
    Q3 = df[column].quantile(0.6)
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR

    upper_array = np.where(df[column] >= upper)[0]
    lower_array = np.where(df[column] <= lower)[0]
    try:
        df.drop(df.index[upper_array], inplace =True)
    except:
        pass
    try:
        df.drop(df.index[lower_array], inplace = True)
    except:
        pass


remove_outliers("BUILD_YEAR")
remove_outliers("LAND_AREA")
remove_outliers("NEAREST_STN_DIST")
remove_outliers("CBD_DIST")
remove_outliers("NEAREST_SCH_DIST")
remove_outliers("FLOOR_AREA")



print(df.shape)

fig, axs = plt.subplots(len(df.columns)-1, figsize=(7, 60))
for n, col in enumerate(df.columns[1:]):
    axs[n].set_title(f"{col} Outliers")
    sns.boxplot(df[col], ax = axs[n])
```
